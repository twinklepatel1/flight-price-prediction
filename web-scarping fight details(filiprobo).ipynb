### Yatra.com

```python
#getting the url
url = "https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=12%2F11%2F2021&hb=0&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=97063523299&version=1.1&viewName=normal"
driver.get(url)
time.sleep(2)
```

```python
#Airline details
airline_name=[]  
try:
    name=driver.find_elements_by_xpath('//span[@class="i-b text ellipsis"]')
    for i in name:
        airline_name.append(i.text)
except NoSuchElementException:
    airline_name.append('-')
#place
source=[]
try:
    Source=driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"]')
    for i in Source:
        source.append(i.text)
except NoSuchElementException:
    source.append('-')
#Source  
Source=[]
for i in range(0,len(source)-1,2):
    Source.append(source[i])
#destination
destination=[]
for i in range(0,len(source)-1,2):
    destination.append(source[i+1])
# total stops
stops=[]
try:
    stop=driver.find_elements_by_xpath('//span[@class="dotted-borderbtm" or @class="cursor-default"]')
    for i in stop:
        stops.append(i.text)
except NoSuchElementException:
    stops.append('-')
#departure time 
depature_time=[]
try:
    time=driver.find_elements_by_xpath('//div[@class="i-b pr"]')
    for i in time:
        depature_time.append(i.text)
except NoSuchElementException:
    depature_time.append('-')
#Arrival time
arrival_time=[]
try:
    time=driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')
    for i in time:
        arrival_time.append(i.text)
except NoSuchElementException:
    arrival_time.append('-')
#duration
duration=[]
try:
    time=driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')
    for i in time:
        duration.append(i.text)
except NoSuchElementException:
    duration.append('-')
#fare
price=[]
try:
    fare=driver.find_elements_by_xpath('//div[@class="pr i-b mr-2"]')
    for i in fare:
        price.append(i.text)
except NoSuchElementException:
    price.append('-')
#Making dataframe
# creating a pandas dataframe
df= pd.DataFrame({'airline_name':airline_name,'date_of_journey':'12/11/2021','Source':Source,'Destination':destination,'depature_time':depature_time,'arrival_time':arrival_time,'duration':duration,'total_stops':stops,'Price':price})
  # show the dataframe
df
```
